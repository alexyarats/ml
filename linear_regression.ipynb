```
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```
diabetes = datasets.load_diabetes()
```

```
x_train, x_valid = diabetes.data[:350], diabetes.data[350:]
y_train, y_valid = diabetes.target[:350], diabetes.target[350:]
```

```
W = np.random.random(10)
b = np.random.random(1)
```

```
# forward pass
def f(x, W, b):
  return x.dot(W) + b
```

```
def L(y_hat, y):
  return np.average(np.square(y_hat - y))
```

```
# backward pass
def dfdw(x_train, y_hat, y):
  print((y_hat - y).shape)
  print((y_hat - y).reshape(y.shape[0], 1).shape)
  print(x_train.shape)
  return 2 * np.average((y_hat - y).reshape(y.shape[0], 1) * x_train, axis=0)
```

```
# backward pass
def dfdb(y_hat, y):
  return 2 * np.average(y_hat - y)
```

```
learning_rate = 0.01
train_loss, valid_loss = [], []
for i in range(10000):
  y_hat = f(x_train, W, b)
  
  t_loss = L(y_hat, y_train)
  v_loss = L(f(x_valid, W, b), y_valid)
  
  train_loss.append(t_loss)
  valid_loss.append(v_loss)
  
  W -= learning_rate * dfdw(x_train, y_hat, y_train)
  b -= learning_rate * dfdb(y_hat, y_train)
  
plt.ylabel("loss")
plt.xlabel("epoch")
plt.title("loss vs. epoch")
plt.tight_layout()
plt.plot(train_loss, label="training")
plt.plot(valid_loss, label="validation")
plt.legend()
```
